```python
def add_features(df1):
    df = df1.copy()
    df['TS_DP'] = df['TS inlet pressure (Mean)'] - df['TS outlet pressure (Mean)']
    df['Pump_to_TS_DP'] = df['Pump outlet pressure (Mean)'] - df['TS inlet pressure (Mean)']
    df['Total_System_DP'] = df['Pump outlet pressure (Mean)'] - df['TS outlet pressure (Mean)']
    return df
```

```python
no_p_dfs = []
p_dfs = []

no_plug_df['Time'] = pd.to_datetime(no_plug_df['Time'].str.replace(',', '.'), format='%H:%M:%S.%f')
no_plug_df = add_features(no_plug_df)
no_p_dfs.append(no_plug_df)

plug_df['Time'] = pd.to_datetime(plug_df['Time'].str.replace(',', '.'), format='%H:%M:%S.%f')
plug_df = add_features(plug_df)
p_dfs.append(plug_df)


for df in raw_no_p_dfs:
    df.drop(columns = ['elapsed_runtime_m', 'elapsed_runtime_s', 'reading_interval_s'], inplace=True)
    df["Time"] = pd.to_datetime(df["Time"], errors="coerce")
    df["Time"] = df["Time"].dt.strftime("%H:%M:%S.%f")
    if 'TS inlet pressure (Mean)' not in df.columns:
        df['TS inlet pressure (Mean)'] = df['TS outlet pressure (Mean)'] + df['Differential pressure (Mean)']
    df = add_features(df)
    no_p_dfs.append(df)

for df in raw_p_dfs:
    df.drop(columns = ['elapsed_runtime_m', 'elapsed_runtime_s', 'reading_interval_s'], inplace=True)
    df["Time"] = pd.to_datetime(df["Time"], errors="coerce")
    df["Time"] = df["Time"].dt.strftime("%H:%M:%S.%f")
    if 'TS inlet pressure (Mean)' not in df.columns:
        df['TS inlet pressure (Mean)'] = df['TS outlet pressure (Mean)'] + df['Differential pressure (Mean)']
    df = add_features(df)
    p_dfs.append(df)    
```

```python
master_column_order = [
    'Time',
    'Flow rate (Mean)',
    'TS outlet pressure (Mean)',
    'TS inlet pressure (Mean)',
    'Pump outlet pressure (Mean)',
    'Temperature TS outlet (Mean)',
    'Tank temperature (Mean)',
    'Temperature TS inlet (Mean)',
    'Bypass temperature (Mean)',
    'TS_DP',
    'Pump_to_TS_DP',
    'Total_System_DP'
]
```

```python
ordered_n_dfs = []
ordered_p_dfs = []

for df in no_p_dfs:
    df_reordered = df[master_column_order].copy()
    ordered_n_dfs.append(df_reordered)
for df in p_dfs:
    df_reordered = df[master_column_order].copy()
    ordered_p_dfs.append(df_reordered)
```

```python
def standardize_time_and_resample(df, interval='1s'):
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S.%f')
    df = df.set_index('Time').sort_index()

    resampled_df = df.resample(interval).mean().interpolate(method='linear')
    
    return resampled_df.reset_index()

standardized_n_dfs = [standardize_time_and_resample(df) for df in ordered_n_dfs]
standardized_p_dfs = [standardize_time_and_resample(df) for df in ordered_p_dfs]
```

```python
def extract_features(df_list, period=12):
    all_features = []
    processed_dfs = []
    t_win = 181 # 1-minute trend window at 1Hz
    
    for df in df_list:
        temp = df.copy()
        
        # 1. Trend of the Total System Pressure
        temp['total_dp_trend'] = STL(temp['Total_System_DP'], period=period, trend=t_win, robust=True).fit().trend
        
        # 2. Pressure Balance (Dimensionless and Flow-Independent)
        # Avoid division by zero with a tiny epsilon
        temp['p_balance'] = temp['TS_DP'] / (temp['Pump_to_TS_DP'] + 1e-6)
        temp['p_balance_trend'] = STL(temp['p_balance'], period=period, trend=t_win, robust=True).fit().trend
        
        # 3. Volatility (High-frequency noise)
        # We take the residual (noise) of the Total DP and look at its rolling standard deviation
        resids = STL(temp['Total_System_DP'], period=period, trend=t_win, robust=True).fit().resid
        temp['volatility'] = resids.rolling(window=30).std() # 30-second jitter window
        
        # 4. Thermal Delta
        temp['temp_delta'] = temp['Temperature TS inlet (Mean)'] - temp['Tank temperature (Mean)']
        
        temp = temp.dropna()
        processed_dfs.append(temp)
        
        # The feature matrix for the Isolation Forest
        features = ['total_dp_trend', 'p_balance_trend', 'volatility', 'temp_delta']
        all_features.append(temp[features])
        
    return pd.concat(all_features), processed_dfs
```

```python
X_train_raw, clean_processed = extract_features(standardized_n_dfs)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_raw)
```

```python
model = IsolationForest(n_estimators=300, contamination=0.001, random_state=42)
model.fit(X_train_scaled)
```

```python
# 3. Extract features from PLUG data
_, plug_processed = extract_features(standardized_p_dfs)

# 4. Score every file
scored_plug_dfs = []
features_list = ['total_dp_trend', 'p_balance_trend', 'volatility', 'temp_delta']

for df in plug_processed:
    temp = df.copy()
    X_test = scaler.transform(temp[features_list])
    
    # decision_function: Positive = Normal, Negative = Anomaly
    temp['anomaly_score'] = model.decision_function(X_test)
    
    # Smooth the score to avoid 'flickering' alarms
    temp['smooth_score'] = temp['anomaly_score'].rolling(window=120).mean()
    scored_plug_dfs.append(temp)
```

```python
# 1. Score the Clean (Normal) Files
scored_clean_dfs = []
for df in clean_processed:
    temp = df.copy()
    X_test = scaler.transform(temp[features_list])
    temp['anomaly_score'] = model.decision_function(X_test)
    # Applying the same 120s smoothing we use for plugs
    temp['smooth_score'] = temp['anomaly_score'].rolling(window=120).mean()
    scored_clean_dfs.append(temp)

# 2. Now calculate the Minima safely
# We use dropna() because the rolling window creates NaNs at the start
min_n = min([df['smooth_score'].dropna().min() for df in scored_clean_dfs])
min_p = min([df['smooth_score'].dropna().min() for df in scored_plug_dfs])

print(f"Deepest Anomaly in Normal Runs: {min_n:.4f}")
print(f"Deepest Anomaly in Plug Runs:   {min_p:.4f}")
```

```python
import matplotlib.pyplot as plt
import math

# 1. Configuration
threshold = -0.0025 # Safely below your normal min of -0.0016
cols = 4
rows = math.ceil(len(scored_plug_dfs) / cols)

fig, axes = plt.subplots(rows, cols, figsize=(20, 4 * rows))
axes = axes.flatten()

for i, df in enumerate(scored_plug_dfs):
    ax = axes[i]
    
    # Logic: 1 if anomaly, 0 if normal
    df['is_alert'] = (df['smooth_score'] < threshold).astype(int)
    
    # Plot the smoothed anomaly score
    ax.plot(df.index, df['smooth_score'], color='blue', linewidth=1.5, label='Anomaly Score')
    
    # Highlight the 'Alert' regions in red
    ax.fill_between(df.index, df['smooth_score'].min(), 0.05, 
                    where=df['is_alert']==1, color='red', alpha=0.3)
    
    # Threshold Line
    ax.axhline(y=threshold, color='black', linestyle='--', alpha=0.6)
    
    ax.set_title(f"Plug Run {i+1}", fontweight='bold')
    ax.set_ylabel("Score")
    ax.set_ylim([min(min_p, -0.01), 0.05]) # Dynamically scale to see the dive
    ax.grid(True, linestyle=':', alpha=0.7)

# Cleanup unused axes
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.suptitle(f"Unsupervised Detection: Pressure-Balance Feature Set", fontsize=18, y=1.01)
plt.show()
```
